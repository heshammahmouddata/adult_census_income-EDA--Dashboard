```python
data.replace("?", pd.NA, inplace=True)
```

```python
# convert data from category to numerical 
data['income'] = data['income'].map({'<=50K': 0, '>50K': 1}).astype('int64')
data
```

```python
# to calc correlation of columns
numeric_columns=data.select_dtypes(np.number)
numeric_columns.columns
```

```python
mode_country = data["native.country"].mode()[0]
# filling null in native_country with mode
data["native.country"] = data["native.country"].fillna(mode_country)
```
